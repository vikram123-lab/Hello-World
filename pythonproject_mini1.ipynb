```python
import random
number = random.randrange(1,100)
guesscheck = "wrong"
print("welcome to game")
guess = 0
while guesscheck == "wrong":
    try:
        response = int(input("enter the number between 1 to 100:"))
    
        #val = int(response)
    except ValueError:
        print("this is not a valid integer")
        continue
    val = int(response)
    guess = guess+1
    if val<number:
        print("lower number")
    elif val>number:
        print("higher number")
    else:
        print("correct number")
        print("you have guessed the correct number in",guess,"times")
        guesscheck = "correct"
print("thanks for playing guess number game")
```
